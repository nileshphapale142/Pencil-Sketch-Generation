```python
from PencilDrawingBySketchAndTone import *
import matplotlib.pyplot as plt
```

```python
input_image = io.imread('./input_images/messi.jpg')
```

```python
pencil_tex = './pencils/pencil0.jpg'
pencil_sketch = gen_pencil_drawing(input_image, rgb=True, w_group=0, pencil_texture_path=pencil_tex,
                               stroke_darkness=2, tone_darkness=1.5)
pencil_sketch = exposure.rescale_intensity(pencil_sketch, in_range=(0,1))

plt.rcParams['figure.figsize'] = [16,12]

plt.subplot(1,2,1)
plt.imshow(input_image)
plt.axis("off")

plt.subplot(1,2,2)
plt.imshow(pencil_sketch)
plt.axis("off")
plt.show()
```
